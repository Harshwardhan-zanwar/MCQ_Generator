````python
from langchain_core.runnables import RunnableSequence, RunnableLambda

# quiz_chain produces an AIMessage. Wrap it to convert into dict.
quiz_to_dict = RunnableLambda(lambda msg: {"quiz": msg.content})

# Now feed that dict into the review prompt
generate_evaluate_chain = RunnableSequence(quiz_chain | quiz_to_dict | review_chain)
````

````python
quiz_to_dict = RunnableLambda(
    lambda msg_and_input: {
        "quiz": msg_and_input[0].content,   # quiz output
        "subject": msg_and_input[1]["subject"]  # original subject
    }
)

generate_evaluate_chain = RunnableSequence(
    quiz_chain.bind(subject=SUBJECT) | quiz_to_dict | review_chain
)
````

````python
from langchain_core.runnables import RunnableSequence, RunnableParallel, RunnableLambda

# Step 1: quiz_chain produces an AIMessage
# Step 2: convert that AIMessage into {"quiz": ...}
quiz_to_dict = RunnableLambda(lambda msg: {"quiz": msg.content})

# Step 3: merge quiz output with original subject
merge_inputs = RunnableParallel({
    "quiz": quiz_to_dict,
    "subject": RunnableLambda(lambda x: x["subject"])
})

# Step 4: feed both into review_chain
generate_evaluate_chain = RunnableSequence(quiz_chain | merge_inputs | review_chain)
````

````python
from langchain_core.runnables import RunnableSequence, RunnableLambda

# Step 1: quiz_chain produces AIMessage
# Step 2: convert AIMessage + original input into dict
quiz_to_dict = RunnableLambda(
    lambda inputs: {
        "quiz": inputs["quiz_output"].content,   # AIMessage content
        "subject": inputs["subject"]             # carry subject forward
    }
)

# Step 3: build pipeline
generate_evaluate_chain = RunnableSequence(
    RunnableLambda(lambda x: {"quiz_output": quiz_chain.invoke(x), **x}) | quiz_to_dict | review_chain
)
````

````python
response = generate_evaluate_chain.invoke({
    "text": TEXT,
    "number": NUMBER,
    "subject": SUBJECT,
    "tone": TONE,
    "response_json": json.dumps(RESPONSE_JSON)
})

print(response)
````

````python
import json

# Step 1: extract the JSON string from the first element's "text"
raw_text = quiz_dict[0]["text"]

# Step 2: isolate the JSON part (between ```json ... ```)
start = raw_text.find("{")
end = raw_text.rfind("}")
quiz_json_str = raw_text[start:end+1]

# Step 3: parse into dict
quiz_data = json.loads(quiz_json_str)

# Step 4: iterate properly
quiz_table_data = []
for key, value in quiz_data.items():
    mcq = value["mcq"]
    options = " | ".join(
        [f"{option}: {option_value}" for option, option_value in value["options"].items()]
    )
    correct = value["correct"]
    quiz_table_data.append({"MCQ": mcq, "Choices": options, "Correct": correct})

print(quiz_table_data)
````
